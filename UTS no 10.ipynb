```python
# UTS no 10
# Menentukan dua transfer function (H1 dan H2) dari masing-masing input terhadap output
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import StateSpace, ss2tf, step

# Matriks sistem MISO
A = np.array([[0, 1],
              [-1, -3]])
B = np.array([[0, 0],
              [4, 4]])            # 2 input
C = np.array([[1, 0]])            # 1 output
D = np.array([[0, 0]])            # D harus (1, 2)

# Buat sistem SISO untuk input ke-1
B1 = B[:, [0]]  # Ambil kolom pertama
D1 = D[:, [0]]
system1 = StateSpace(A, B1, C, D1)

# Buat sistem SISO untuk input ke-2
B2 = B[:, [1]]  # Ambil kolom kedua
D2 = D[:, [1]]
system2 = StateSpace(A, B2, C, D2)

# Simulasi step response
t = np.linspace(0, 10, 500)
t1, y1 = step(system1, T=t)
t2, y2 = step(system2, T=t)

# Plot hasil
plt.figure(figsize=(8, 4))
plt.plot(t1, y1, label='Step Response H1 (Input 1)')
plt.plot(t2, y2, label='Step Response H2 (Input 2)', linestyle='--')
plt.xlabel('Waktu (s)')
plt.ylabel('Keluaran y(t)')
plt.title('Perbandingan Step Response MISO (SISO per input)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()


```
