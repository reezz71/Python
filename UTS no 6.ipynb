```python
# UTS no 6
import control
import matplotlib.pyplot as plt
import numpy as np

# Definisikan matriks state-space
A = [[0, 1],
     [0, -1]]
B = [[0],
     [1]]
C = [[1, 0]]
D = [[0]]

# Buat sistem state-space
sys = control.ss(A, B, C, D)

# Menampilkan sistem
print("State-space model:")
print("A =", np.array(A))
print("B =", np.array(B))
print("C =", np.array(C))
print("D =", np.array(D))

# Simulasi step response
time = np.linspace(0, 10, 1000)
t, y = control.step_response(sys, T=time)

# Plot hasil
plt.figure()
plt.plot(t, y)
plt.title("Step Response dari Sistem")
plt.xlabel("Waktu (s)")
plt.ylabel("Output y(t)")
plt.grid(True)
plt.show()
```
