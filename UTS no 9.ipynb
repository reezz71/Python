```python
# UTS no 9
# Menampilkan transfer function H(s) dan plot step response-nya.
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import ss2tf, StateSpace, step

# Definisi matriks state-space
A = np.array([[0, 1],
              [-2, -3]])
B = np.array([[0],
              [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

# Konversi ke fungsi alih (transfer function)
num, den = ss2tf(A, B, C, D)
print("Fungsi alih H(s):")
print("Numerator:", num)
print("Denominator:", den)

# Buat sistem state-space
system = StateSpace(A, B, C, D)

# Simulasi step response
t, y = step(system)

# Plot hasil
plt.figure(figsize=(8, 4))
plt.plot(t, y, label='Step Response')
plt.xlabel('Waktu (s)')
plt.ylabel('Keluaran y(t)')
plt.title('Step Response Sistem SISO')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
