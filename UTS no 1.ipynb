```python
# UTS No. 1 - Analisis Step Response Sistem Kontrol

# Import library yang dibutuhkan
import control as ct                # Library untuk analisis sistem kontrol
import numpy as np                  # Library untuk operasi numerik
import matplotlib.pyplot as plt     # Library untuk visualisasi grafik

# Parameter sistem
K = 4       # Gain sistem
T = 3       # Konstanta waktu
U = 2       # Delay (tidak digunakan dalam transfer function ini)

num = np.array([K])        # Numerator dari fungsi transfer
den = np.array([T, U, 1])     # Denominator dari fungsi transfer

# Membuat fungsi transfer
H = ct.tf(num, den)
print('H(s) =', H)

# Menghitung dan mem-plot respons terhadap input langkah
t, y = ct.step_response(H)

plt.plot(t, y)
plt.title("Step Response")
plt.grid()
```
