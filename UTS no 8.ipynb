```python
# UTS no 8
import numpy as np

# Matriks sistem kontinu
A = np.array([[0, -1],
              [1, -3]])
B = np.array([[0],
              [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

# Waktu sampling
Ts = 0.1

# Diskritisasi dengan metode Euler
Ad = np.eye(2) + Ts * A
Bd = Ts * B
Cd = C
Dd = D

# Tampilkan hasil
print("Ad (A diskrit):\n", Ad)
print("Bd (B diskrit):\n", Bd)
print("Cd (C diskrit):\n", Cd)
print("Dd (D diskrit):\n", Dd)
# Sistem diskrit ini merepresentasikan dinamika sistem kontinu yang telah disampling setiap 0.1 detik.
# Matriks Ad menunjukkan bagaimana keadaan berubah dari satu waktu ke waktu berikutnya.
```
