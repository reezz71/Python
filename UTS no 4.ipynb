```python
# UTS no 4
import control as ct # untuk membuat dan menganalisis sistem kontrol seperti transfer function, step response
import matplotlib.pyplot as plt # untuk membuat grafik visualisasi
import numpy as np # untuk operasi numerik seperti membuat array

# Nilai gain yang akan diuji
K_values = [1, 2, 5, 10] # Kita akan membuat sistem dengan masing-masing nilai K dan membandingkan responsnya

# Waktu simulasi
t = np.linspace(0, 50, 500)

# Plot semua step response dalam satu grafik
plt.figure() # untuk menampung semua grafik step response dalam satu tampilan
for K in K_values:
    num = [K]
    den = [10, 1]
    H = ct.tf(num, den) # membuat transfer fungsi
    print ('H(s)=', H) # bentuk tf
    y, t_out = ct.step_response(H, T=t)
    plt.plot(t_out, y, label=f"K = {K}")

# Format grafik
plt.title("Pengaruh Nilai K terhadap Step Response Sistem Orde Satu")
plt.xlabel("Waktu (s)")
plt.ylabel("Output")
plt.grid(True) # memberikan garis untuk memudahkan dalam membaca grafik
plt.legend()
plt.show() # menampilkan grafik akhir 
```
