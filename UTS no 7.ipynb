```python
# UTS no 7
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import StateSpace, step

# Definisi matriks state-space
A = np.array([[0, -1],
              [1, -3]])
B = np.array([[0],
              [1]])
C = np.array([[1, 0]])
D = np.array([[0]])

# Buat sistem state-space
system = StateSpace(A, B, C, D)

# Simulasi step response
t, y = step(system)

# Plot hasil
plt.figure(figsize=(8, 4))
plt.plot(t, y, label='Step Response')
plt.xlabel('Waktu (s)')
plt.ylabel('Keluaran y(t)')
plt.title('Step Response Sistem')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

from scipy.signal import ss2tf

# Konversi ke transfer function
num, den = ss2tf(A, B, C, D)

print("Numerator:", num)
print("Denominator:", den)

# Sistem ini adalah orde dua karena penyebutnya s^2+3s+1
# Karakteristik dinamikanya ditentukan oleh posisi pole: jika kompleks konjugat → sistem osilatif; jika real → sistem overdamped

```
