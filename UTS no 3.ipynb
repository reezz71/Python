```python
# UTS no 3
import control as ct
import matplotlib.pyplot as plt

# Parameter delay
delay = 2      # Delay time (s)
orde = 3    # Orde aproksimasi

# Mendapatkan Pade Approximation dari e^(-s*tau)
num, den = ct.pade(delay, orde)
delay_tf = ct.tf(num, den)

print("Transfer Function Pade Approximation (e^{-2s}):")
print(delay_tf)

# Plot step response dari fungsi delay
t, y = ct.step_response(delay_tf)

plt.plot(t, y)
plt.title("Step Response of Pade Approximation (Ï„ = 2s, N = 3)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
